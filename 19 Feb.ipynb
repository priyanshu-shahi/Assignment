Flask is a lightweight web framework for Python, designed to make it easy to develop web applications quickly and with minimal code.

Some advantages of Flask include-
  1. Lightweight: Flask is minimalistic and doesn't impose any dependencies or project structure, allowing developers to 
     start small and scale up as needed.
  2. Flexibility: Flask gives developers the freedom to choose components and libraries they prefer, making it highly
     customizable.
  3. Extensibility: Flask supports extensions that add extra functionality, allowing developers to easily integrate 
     features like authentication, database integration, and more.
  4. Easy to learn: Flask has a simple and intuitive syntax, making it accessible for beginners and allowing for rapid 
     development.
  5. Built-in development server: Flask comes with a built-in development server, making it easy to test and debug
     applications during development.
  6. RESTful request dispatching: Flask supports RESTful request dispatching, making it well-suited for building RESTful
     APIs.
  7. Active community: Flask has a large and active community of developers, providing ample resources, tutorials, and
     plugins to support development.

App routing in Flask refers to the mechanism of mapping URLs to specific functions or views within a Flask application. It is achieved using the `@app.route()` decorator, where you specify the URL endpoint and the associated function to handle requests to that endpoint. 
We use app routes to define the different pages or functionalities of a web application. Each route corresponds to a specific URL, and when a user accesses that URL, Flask knows which function to call to generate the appropriate response. This helps in organizing the codebase and making it easier to maintain and extend the application.

Required code is-

```python
! pip install flask
from flask import Flask
app = Flask(__name__)
@app.route("/Welcome")
def Welcome():
    return "Welcome to ABC Corporation"

@app.route("/")
def Details():
    details= "<h1>Company Name: ABC Corporation</h1><h1>Location: India</h1><h1>Contact Detail: 999-999-9999</h1>"
    return details

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

In Flask, the function used for URL building is url_for(). This function generates a URL for the specified endpoint, taking into account any variable parts of the URL.

Required code is-

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the index page. <a href="' + url_for('about') + '">About</a>'

@app.route('/about')
def about():
    return 'This is the about page. <a href="' + url_for('index') + '">Index</a>'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
